```python
X = ranked.drop(['Fish', 'Stockid'], axis=1)

X = StandardScaler().fit_transform(X)
X.shape
```

```python
db = DBSCAN(eps=0.18, min_samples=2).fit(X)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
labels = db.labels_
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
labels
```

```python
%matplotlib inline
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = ['lightgreen', 'red', 'orange']
plt.figure(figsize=(8,7))
plt.axvline(x=0.0, color='lightgray')
plt.axhline(y=0.0, color = 'lightgray')

for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=8)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=6)

plt.title('health of fish populations')
plt.xlabel('exploitable population (MT)')
plt.ylabel('fishing rate')
# plt.legend()
plt.ylim(-1,5)
plt.xlim(-2, 8)

sns.despine()
plt.show()
```

## K Means

```python
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans, KMeans
from sklearn.metrics.pairwise import pairwise_distances_argmin
```

```python
k_means = KMeans(init='k-means++', n_clusters=3, n_init=8)
```

```python
fit = k_means.fit(X)
```

```python
labels = k_means.fit_predict(X)
```

```python
%matplotlib inline
# Black removed and is used for noise instead.
unique_labels = set(labels)
colors = plt.cm.Spectral(np.linspace(0, 1, len(unique_labels)))
plt.figure(figsize=(10,10))
for k, col in zip(unique_labels, colors):
    if k == -1:
        # Black used for noise.
        col = 'k'

    class_member_mask = (labels == k)

    xy = X[class_member_mask & core_samples_mask]
    
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='k', markersize=14)

    xy = X[class_member_mask & ~core_samples_mask]
    plt.plot(xy[:, 0], xy[:, 1], 'o', markerfacecolor=col,
             markeredgecolor='black', markersize=6)

plt.title('Estimated number of clusters: %d' % n_clusters_)
sns.despine()
plt.show()
```
